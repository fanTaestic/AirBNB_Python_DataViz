**Task 4: DATA CLEANING**

```python
data.isnull().sum()

data.dropna(inplace=True)     #dropping all missing values rows

#data.fillna()------to fill value
```

```python
data.duplicated().sum()  #to see the duplicate values in the dataset
```

```python
data[data.duplicated()]
```

```python
# deleting all duplicated records
data.drop_duplicates(inplace=True)
```

```python
#type casting
# changing data types

data.dtypes

# change the datatype
data['id']=data['id'].astype(object)
data.dtypes

data['host_id']=data['host_id'].astype(object)
data.dtypes
```

## EDA

**Task 5:  DATA ANALYSIS**

**Univariate Analysis**


```python
#identify outliers in price

df = data[data['price'] < 1500]
sns.boxplot(data=df, x='price')
```

```python
# Price Distribution 
data['price']
```

```python
#price distribution histogram 

plt.figure(figsize=(8, 5))   # to give a fixed size
sns.histplot(data=df, x='price', bins=100)
plt.title('Price Distribution')
plt.ylabel("Frequency")    # change the y label name 
plt.show()
```

```python
df.columns
```

```python
df.dtypes
```

```python
# Availibility Distribution

plt.figure(figsize=(6, 3))   # to give a fixed size
sns.histplot(data=df, x='availability_365')
plt.title('Availability_365 Distribution')
plt.ylabel("Frequency")    # change the y label name 
plt.show()
```

```python
data.dtypes
```

```python
# Average price in each neighbourhood

df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
df.groupby(by='neighbourhood_group')['Price per Bed'].mean()
```

**FEATURE ENGINEERING**

```python
# ['Price per bed']

df['Price per Bed']= df['price']/df['beds']
df.head()
```

**Bi Variable Analysis** to analyse one column relation with other column

```python
df.columns
```

```python
# Price Dependency on Neighbourhood

sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
```

```python
# Analyse no. of reiews with Price

plt.figure(figsize=(8, 5))
plt.title("Locality and Review Dependency")
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

```python
df.dtypes
```

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
```

```python
# Geographic distribution of AirBNB Listing

plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title("Geographic Distribution of AirBnB Listing")
plt.show()
```

```python
# Correlation using number columns, heat maps(finding subsets)

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr
plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True)
```
